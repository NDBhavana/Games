```python
import random
m={}
p1=0
p2=0
l=['1','2','3','4','5','6','7','8','9']
print("CSP: Tic Tac Toe\nPositions on board\n|1|2|3|\n|4|5|6|\n|7|8|9|\n")
def evalrow(m):
    if((m['1']==m['2']==m['3'] or m['4']==m['5']==m['6'] or m['7']==m['8']==m['9']) and((m['1']=="X"or m['4']=="X" or m['7']=="X")or(m['1']=="O" or m['4']=="O" or m['7']=="O"))):
        if((m['1']=="X" and m['1']==m['2']==m['3']) or (m['4']=="X" and m['4']==m['5']==m['6']) or (m['7']=="X" and m['7']==m['8']==m['9'])):
            return 1
        elif((m['1']=="O" and m['1']==m['2']==m['3']) or (m['4']=="O" and m['4']==m['5']==m['6']) or (m['7']=="O" and m['7']==m['8']==m['9'])):
            return 2
        elif(m['1']==" " or m['4']==" " or m['7']==" "):
            return 0
    return 0
def evaldiag(m):
    if(m['1']==m['5']==m['9'] or m['3']==m['5']==m['7']):
        if((m['1']=="X" and m['1']==m['5']==m['9'] ) or (m['3']=="X" and m['3']==m['5']==m['7'])):
            return 1
        elif((m['1']=="O" and m['1']==m['5']==m['9'] ) or (m['3']=="O" and m['3']==m['5']==m['7'])):
            return 2
        elif(m['1']==" " or m['3']==" "):
            return 0
    return 0
def evalcol(m):
    if(m['1']==m['4']==m['7'] or m['2']==m['5']==m['8'] or m['3']==m['6']==m['9']):
        if((m['1']=="X" and m['1']==m['4']==m['7']) or (m['2']=="X" and m['2']==m['5']==m['8']) or(m['3']=="X" and m['3']==m['6']==m['9']) ):
            return 1
        elif((m['1']=="O" and m['1']==m['4']==m['7']) or (m['2']=="O" and m['2']==m['5']==m['8']) or(m['3']=="O" and m['3']==m['6']==m['9'])):
            return 2
        elif(m['1']==" " or m['2']==" " or m['3']==" "):
            return 0
    return 0
def winner(u,v,w):
    if(u):
        if(u==1):
            print("Player 1 won")
        elif(u==2):
            print("Player 2 won")
    elif(v):
        if(v==1):
            print("Player 1 won")
        elif(v==2):
            print("Player 2 won")
    elif(w):
        if(w==1):
            print("Player 1 won")
        elif(w==2):
            print("Player 2 won")
    else:
        print("TIE")
for i in l:
    m[i]=" "
while(l!=[] and not(evalrow(m)>0 or evalcol(m)>0 or evaldiag(m)>0) ):
    print("Player 1")
    k=random.choice(l)
    m[k]="X"
    p1+=1
    l.remove(k)
    for i in range(9):
        print("|",m[str(i+1)],"|",end="")
        if(i==2 or i==5 or i==8):
            print()
    if(l==[]):
        break
    if((evalrow(m)>0 or evalcol(m)>0 or evaldiag(m)>0)):
        break
    print("Player 2")
    k=" "
    while(k not in l):
        k=str(input("Choose available move:"))
    m[k]="O"
    p2+=1
    l.remove(k)
    for i in range(9):
        print("|",m[str(i+1)],"|",end="")
        if(i==2 or i==5 or i==8):
            print()
winner(evalrow(m),evalcol(m),evaldiag(m))
```
